```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.seq_len = seq_len
        self.random_dim = random_dim
        self.model = nn.Sequential(
            nn.ConvTranspose1d(self.random_dim, self.seq_len // 4, 1, bias=False),
            nn.BatchNorm1d(self.seq_len // 4),
            nn.ReLU(True),

            nn.ConvTranspose1d(self.seq_len // 4, self.seq_len // 2, 1, bias=False),
            nn.BatchNorm1d(self.seq_len // 2),
            nn.ReLU(True),
            
            nn.ConvTranspose1d(self.seq_len // 2, self.seq_len, 1, bias=False)
      )
    def forward(self, z):
        z = z.unsqueeze(-1)
        output = self.model(z)
        return output.squeeze(-1)
```

```python
# Functions to calculate parameters
def slope(mas):
    slopes = []
    for i in range(mas.shape[0]):
        m = max(mas[i])
        idx = list(mas[i]).index(m) #index of a max element
        s = 0
        for j in range(idx - 50, idx - 20):
            s += mas[i][j]
        avg1 = s / 30 #1st y-coordinate

        s = 0
        for j in range(idx + 1000, idx + 2000):
            s += mas[i][j]
        avg2 = s / 1000 #2nd y-coordinate

        x = (idx + 1500) - (idx - 35)
        y = abs(avg2 - avg1)
        slopes.append(x / y)
    return np.array(slopes)

def drift(mas):
    drifttimes = []
    for j in range(mas.shape[0]):
        m = max(mas[j])
        idx = list(mas[j]).index(m) #index of a max element
        for i in range(idx):
            if (mas[j][i] - 0.9*m) <= 0.01:
                k90 = i
            if (mas[j][i] - 0.1*m) <= 0.01:
                k10 = i
        drifttimes.append(k90 - k10)
    drifttimes = np.array(drifttimes)
    return drifttimes
```

```python
# parameters of the real pulses
# energy amplitude
max_real = np.max(real, axis = 1) - np.min(real, axis = 1)
# current amplitude
real_c = np.diff(real, axis=1)
max_real_c = np.max(real_c, axis = 1) - np.min(real_c, axis = 1)
# shift for tail slope calculation
real = real - np.min(real, axis=1).reshape(real.shape[0], 1)
# tail slope
sl_r = slope(real)  
# normalization for drift time calculation
real = (real - np.min(real, axis=1).reshape(real.shape[0], 1)) / (np.max(real, axis=1) - np.min(real, axis=1)).reshape(real.shape[0], 1)
# drift time
rr = drift(real)
#############
# avoid other outliers for better IOU calculation
rr = rr[(rr>20) & (rr<50)]
sl_r = sl_r[sl_r<15]
max_real = max_real[(max_real>1200) & (max_real<1300)]
max_real_c = max_real_c[(max_real_c>50) & (max_real_c<250)]
```

```python
def calculate_iou(h1,h2, bins=50):
    '''
    Calculate the histogram intersection over union (IOU)
    '''
    rg = np.histogram(np.hstack((h1, h2)), bins=bins)[1]
    
    h1 = np.array(h1)
    h2 = np.array(h2)
    
    count, _ = np.histogram(h1,bins=rg,density=True)
    count2, _ = np.histogram(h2,bins=rg,density=True)
    
    intersection = np.sum(np.minimum(count,count2))
    union = np.sum(np.maximum(count,count2))
        
    return intersection/union*100.0
```

### Training the model with different random seeds

```python
iou_en = []
iou_cur = []
iou_slope = []
iou_drift = []
import time

for i in range(1, 701):
    print('Iteration ', i)
    G = Generator()
    file = 'models/generator_pca_' + str(i) + '.pth'
    G.load_state_dict(torch.load(file, map_location=torch.device('cpu')))
    noise = torch.randn(2000, random_dim)
    
    scaler = StandardScaler()
    pca = PCA(n_components=seq_len)
    scaler = pickle.load(open('scaler_pca.pkl','rb'))
    pca = pickle.load(open('pca_pca.pkl','rb'))

    fake = G(noise).detach()
    fake = pca.inverse_transform(fake)
    fake = scaler.inverse_transform(fake)
    
    
    #energy amplitude of the fake pulses
    print('\tEnergy processing')
    max_fake = np.max(fake, axis = 1) - np.min(fake, axis = 1)
        
    #current amplitude of the fake pulses
    print('\tCurrent processing')
    fake_c = np.diff(fake, axis=1)
    max_fake_c = np.max(fake_c, axis = 1) - np.min(fake_c, axis = 1)
    
    #tail slope  of the fake pulses
    print('\tSlope processing')
    fake = fake - np.min(fake, axis=1).reshape(fake.shape[0], 1)
    sl_f = slope(fake)
    
    #drift time of the fake pulses
    print('\tDrift processing')
    fake = (fake - np.min(fake, axis=1).reshape(fake.shape[0], 1)) / (np.max(fake, axis=1) - np.min(fake, axis=1)).reshape(fake.shape[0], 1)
    rf = drift(fake)
    
    # Same cuts to avoid outliers
    rf = rf[(rf>20) & (rf<50)]
    sl_f = sl_f[sl_f<15]
    max_fake = max_fake[(max_fake>1200) & (max_fake<1300)]
    max_fake_c = max_fake_c[(max_fake_c>50) & (max_fake_c<250)]
    ####
    
    # IOU calculation
    print('\tIoU calculation')
    iou_en.append(calculate_iou(max_real, max_fake))
    iou_cur.append(calculate_iou(max_real_c, max_fake_c))
    iou_slope.append(calculate_iou(sl_r, sl_f))
    iou_drift.append(calculate_iou(rr, rf))
    
```

```python
iou_en = np.array(iou_en)
iou_cur = np.array(iou_cur)
iou_slope = np.array(iou_slope)
iou_drift = np.array(iou_drift)
```
